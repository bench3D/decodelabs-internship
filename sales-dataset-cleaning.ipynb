```python
# Inspecting the key columns
# We open a parenthesis to start a "chain" across multiple lines
# CouponCode has 309 missing values, however not all products have coupon cddes - we wil replce with a placeholder "No coupon"
# convert date to datetime type
# convert OrderID, CustomerID, properly to string type
# convert OrderStatus, ReferralSource to category type
df_clean = (
    pd.read_csv(path_1)
    .assign(
        # 1. Fill missing CouponCode with placeholder
        CouponCode = lambda x: 
        x["CouponCode"].fillna("No coupon"),
        
        # 2. Safe Date conversion
        Date = lambda x: 
        pd.to_datetime(x["Date"], errors="coerce"),
        
        # 3. Ensure IDs are strings
        OrderID = lambda x: 
        x["OrderID"].astype(str),
        CustomerID = lambda x: 
        x["CustomerID"].astype(str),
        
        # 4. Convert repetitive text fields to categories
        OrderStatus = lambda x: 
        x["OrderStatus"].astype("category"),
        ReferralSource = lambda x: 
        x["ReferralSource"].astype("category")
    )
)

# Quickly inspect the new dtypes and missing values to verify!
print(df_clean.info())
```

```python
df_clean.describe()
```

```python
# Find all string/object columns
string_cols = df_clean.select_dtypes(include=['str', 'object', 'category']).columns

# Count how many rows have leading/trailing spaces in each column
for col in string_cols:
    space_count = df_clean[col].astype(str).str.contains(r'^\s+|\s+$', regex=True).sum()
    if space_count > 0:
        print(f"Column '{col}' has {space_count} rows with leading/trailing spaces.")
    else:
        print(f"✅ '{col}' is perfectly clean!")
```

```python
# Calculate structural distributions
skewness = df_clean[['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']].skew()

print("--- Skewness Coefficients ---")
print(skewness)
print("\nAnalytical Note: A value > 1 indicates highly right-skewed data, meaning outliers pull the mean away from the truth.")
```
