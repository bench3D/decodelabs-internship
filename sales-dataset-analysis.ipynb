```python
# Inspecting the key columns
# We open a parenthesis to start a "chain" across multiple lines
# CouponCode has 309 missing values, however not all products have coupon cddes - we wil replce with a placeholder "No coupon"
# convert date to datetime type
# convert OrderID, CustomerID, properly to string type
# convert OrderStatus, ReferralSource to category type
df_clean = (
    pd.read_csv(path_1)
    .assign(
        # 1. Fill missing CouponCode with placeholder
        CouponCode = lambda x: 
        x["CouponCode"].fillna("No coupon"),
        
        # 2. Safe Date conversion
        Date = lambda x: 
        pd.to_datetime(x["Date"], errors="coerce"),
        
        # 3. Ensure IDs are strings
        OrderID = lambda x: 
        x["OrderID"].astype(str),
        CustomerID = lambda x: 
        x["CustomerID"].astype(str),
        
        # 4. Convert repetitive text fields to categories
        OrderStatus = lambda x: 
        x["OrderStatus"].astype("category"),
        ReferralSource = lambda x: 
        x["ReferralSource"].astype("category")
    )
)

# Quickly inspect the new dtypes and missing values to verify!
print(df_clean.info())
```

```python
# Find all string/object columns
string_cols = df_clean.select_dtypes(include=['str', 'object', 'category']).columns

# Count how many rows have leading/trailing spaces in each column
for col in string_cols:
    space_count = df_clean[col].astype(str).str.contains(r'^\s+|\s+$', regex=True).sum()
    if space_count > 0:
        print(f"Column '{col}' has {space_count} rows with leading/trailing spaces.")
    else:
        print(f"✅ '{col}' is perfectly clean!")
```

```python
# Calculate structural distributions
skewness = df_clean[['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']].skew()

print("--- Skewness Coefficients ---")
print(skewness)
print("\nAnalytical Note: A value > 1 indicates highly right-skewed data, meaning outliers pull the mean away from the truth.")
```

```python
# Calculate IQR for TotalPrice
Q1 = df_clean['TotalPrice'].quantile(0.25)
Q3 = df_clean['TotalPrice'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Isolate the Outliers (VIP High-Value Transactions)
vip_orders = df_clean[df_clean['TotalPrice'] > upper_bound]

print(f"Outlier Threshold: Orders above ${upper_bound:.2f}")
print(f"Total VIP Signal Rows Found: {len(vip_orders)}")
```

```python
# Run Pearson Correlation
correlation_matrix = df_clean[['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']].corr(method='pearson')

print("--- Pearson Correlation Matrix (r) ---")
print(correlation_matrix)
```

```python
import sqlite3

# 1. Create an in-memory database connection (exists only in your RAM)
conn = sqlite3.connect(':memory:')

# 2. Push your cleaned dataframe into an SQL table named 'sales'
df_clean.to_sql('sales', conn, index=False, if_exists='replace')

# 3. Create a helper function to run queries
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
# Query 1: High-Value Inventory Performance
query_1 = """
SELECT 
    Product, 
    UnitPrice, 
    Quantity, 
    TotalPrice
FROM sales
WHERE UnitPrice > 350.00
ORDER BY TotalPrice DESC;
"""
run_query(query_1)
```

```python
#Query 2: Revenue and Transaction Count by Payment Method
query_2 = """
SELECT 
    PaymentMethod,
    COUNT(*) AS total_transactions,            
    SUM(TotalPrice) AS total_revenue,          
    AVG(TotalPrice) AS average_order_value     
FROM sales
GROUP BY PaymentMethod
ORDER BY total_revenue DESC;
"""
run_query(query_2)
```

```python
# Query 3: Promotional Coupon Impact
query_3 = """
SELECT 
    CouponCode,
    COUNT(*) AS usage_count,
    SUM(TotalPrice) AS promotional_revenue
FROM sales
WHERE CouponCode IS NOT NULL AND CouponCode <> 'No coupon'
GROUP BY CouponCode
ORDER BY usage_count DESC;
"""
run_query(query_3)
```

```python
# Pipeline Health
query_4="""
SELECT 
    OrderStatus,
    COUNT(*) AS order_volume,
    SUM(TotalPrice) AS tied_up_revenue
FROM sales
GROUP BY OrderStatus
HAVING SUM(TotalPrice) > 50000.00 -- Rejection funnel for low-value pipelines
ORDER BY tied_up_revenue DESC;
"""
run_query(query_4)
```
